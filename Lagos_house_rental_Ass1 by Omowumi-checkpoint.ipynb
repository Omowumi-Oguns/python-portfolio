```python
# import the necessary libraries
import numpy as np
import pandas as pd

# import visual python libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
lagos_houses = pd.read_csv(r'C:\Users\HP\Downloads\lagos-house-prices-raw.csv')
lagos_houses
```

```python
# To view the dataset of self contained apartment
self_con = lagos_houses[lagos_houses['Property_Type'] == 'Self contain']
self_con
```

```python
# To check the cheapest location and property type with 2 bedrooms and 2 toilets excluding self contained

# Firstly, we filter the data for 2 bedroom and 2 toilet excluding self contained

filtered_data = lagos_houses[(lagos_houses['Property_Type'] != 'Self contain') & (lagos_houses ['bed'] == 2) & (lagos_houses['toilet'] == 2)]

cheapest_price = filtered_data.groupby(['location', 'Property_Type']).price.min().unstack('location')

cheapest_price   
```

```python
#To visualize it
cheapest_price.plot.bar(figsize = (15,5))
plt.ticklabel_format(style = 'plain', axis ='y')
plt.show()
```

### Observation
- It is observed that miniflat and flat apartment located in Ikorodu has the cheapest price with 2 bedroom and 2 toilet
- It is also observed that miniflat and flat property type with 2 bedroom and 2 toilet  are located in all the 7 districts

```python
# To get the average price by location and property type
avg_price = lagos_houses.groupby(['location', 'Property_Type']).price.mean().astype('int').unstack('Property_Type')
avg_price
```

```python
# Visualization
avg_price.plot.bar(figsize = (18,5))
plt.ticklabel_format(style = 'plain', axis ='y')
plt.show()
```

### Observation
- We can see that houses in Ikorodu are lesser

```python
# summary statistics per security doors and price
lagos_houses.groupby('Security_Doors').price.describe().astype('int')
```

```python
# View distribution of prices by security door
sns.boxplot(x = lagos_houses.Security_Doors, y = lagos_houses.price, data = lagos_houses)
plt.show()
```

### Observation 
- This shows that houses with security doors are more expensive

```python
# To view the dataset of houses with electricity by location
Electricity_loc = lagos_houses['Electricity'].groupby(lagos_houses.location).value_counts()
Electricity_loc
```

```python
# create a visualization 
plt.figure(figsize = (10, 5))
plt.title('Location with electricity')
sns.countplot(x = 'location', data = lagos_houses, hue = 'Electricity')
plt.xlabel('location')
plt.ylabel('Electricity')
```

### Observation
- It is observed that Yaba has the count of Electricity by location

```python
#  To know location with gym service
Gym_loc = lagos_houses['Gym'].groupby(lagos_houses.location).value_counts()
Gym_loc
```

```python
# create a visualization 
plt.figure(figsize = (10, 5))
plt.title('Count of gym facilities by Location')
sns.countplot(x = 'location', data = lagos_houses, hue = 'Gym')
plt.xlabel('Locations')
plt.ylabel('Gym')
```

```python
# To get the average price of penthouse flat per location

pflat_loc = lagos_houses[lagos_houses['Property_Type'] == 'Penthouse flat']
pflat_by_location = pflat_loc.groupby('location').price.mean().astype('int')
pflat_by_location
```

```python
pflat_by_location.plot(kind = 'bar', title = 'Average price of Penthouse Flat by location')
```
