```python
# visualizing missing values
plt.figure(figsize = (10, 5))
plt.title('Missing Values')
sns.heatmap(train.isnull(), cbar = True, cmap = 'magma_r')
```

```python
# create a function that properly labels survived
def survive(x):
    if x == 1:
        return 'Survived'
    else:
        return 'Died'
    
# create a new column
train['survived_label'] = train['Survived'].apply(survive)
```

```python
train.head()
```

```python
# create a visual for survived
plt.figure(figsize = (10,5))
plt.title('Passengers survival on the Titanic')
plt.xlabel('Survived')
plt.ylabel('Count of Passengers')
sns.countplot(x = 'survived_label', data = train)
plt.show()
```

```python
# percentage of survived and died
survived_percent = train['survived_label'].value_counts() / train.shape[0] * 100
print(survived_percent)
```
