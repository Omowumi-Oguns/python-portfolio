### Predicting Lagos House Prices
It is a regression task. In this task we will be looking at the following Algorithm:
- Decision Tree
- Random forest
- Linear regression

```python
lagos_houses.head()
```

### Select Target
- The target is what you want to predict. In this exercise we want to predict house prices.

```python
# we usually tend to name the target y 
y = lagos_houses.price
y.head()
```

### Feature Selection
Normally you don't need to use all columns in your dataframe to predict your target.
- Feature selection is where you choose the features that you will use to train your model.
We tend to call the training data X

```python
# view column names
lagos_houses.columns
```

```python
# create a list of features, for now they are just numerical columns. Remember that price is your target and shouldn't be in your
feature_columns = [ 'bed', 'bath', 'toilet',
       'Parking_Space', 'Security', 'Electricity', 'Furnished',
       'Security_Doors', 'CCTV', 'Pool', 'Gym', 'BQ']
```

```python
# create x by filtering the housing data with selected features
X = lagos_houses[feature_columns]
X.head()
```

### your first machine learning model 

```python
# import the model from scikit 
from sklearn.tree import DecisionTreeRegressor
```

```python
# create an instance of the model
lagos_model = DecisionTreeRegressor(random_state = 42)
```

```python
# fit your model to your data, i.e train your model
lagos_model.fit(X, y)
```

```python
# predict house price
preds = lagos_model.predict(X)
```

```python
print('Actual prices: ', y.head().tolist())
print('Predicted prices: ', preds[:5])
```

```python
'''it appears that our predictions are close to the actual prices. we can use performance statistical metrics to evaluate the model. One such metric is mean absolute error'''
```

### Evaluate Model with MAE(MEAN ABSOLUTE ERROR)
- If the actual price of a house is N750,000 but our model predicts N800,000 the difference is N50,000

statistically, we can that ERROR OR RESIDUAL
- Mean Absolute Error is the mean of all such errors in the range of the predicted values.

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y, preds)
print(mae)
```

```python
# Mean squared error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y, preds)
print(mse)
```

```python
# R squared
from sklearn.metrics import r2_score
score = r2_score(y, preds)
print(score)
```
